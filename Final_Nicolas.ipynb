# Creating new dataset of X and Y

```python
df2= df[['student_Class','Result'
       ,'Term 1 Final Math Average','Term 1 Final English Average',
         'Age at begining of year/monthes','year in School','Math Teacher weekly infraction avg',
         'Missing Exams Term 1 Math','Missing Exams Term 1 English','Missing Exams Term 1 Science','Term 1 # days absent',
         'Term 1 # infractions','AvgMathPeriodic','AvgEnglishPeriodic','AvgSciencePeriodic','AvgMathAMS']]
       #,'School','Nationality','Math Teacher']]
df2.dropna(inplace=True)
X = df2.drop('Result', axis=1)
y = df2.Result
```

```python
model1 = lm.LogisticRegression(C=1e50)
model1.fit(X, y)
```

```python
list(zip(X.columns, np.exp(model1.coef_[0,:])))
```

```python
five_fold_cv = ms.StratifiedKFold(n_splits=5, shuffle=True)
```

```python
accuracies = ms.cross_val_score(lm.LogisticRegression(C=1e50), X, y, scoring='accuracy', cv=five_fold_cv)
np.mean(accuracies)
```

```python
aucs = ms.cross_val_score(lm.LogisticRegression(C=1e50), X, y, scoring='roc_auc', cv=five_fold_cv)
np.mean(aucs)
```

```python
model2 = lm.LogisticRegressionCV(Cs=10, cv=five_fold_cv, scoring='roc_auc')
model2.fit(X, y)
```

```python
list(zip(X.columns, np.exp(model2.coef_[0,:])))
```

```python
metrics.confusion_matrix(y, model2.predict(X))
```
